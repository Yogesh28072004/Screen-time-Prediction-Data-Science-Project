```python
#histogram of hours of screen on time per day

plt.figure(figsize=(10, 6))
sns.histplot(df['Screen On Time (hours/day)'], bins=20, kde=True)
plt.title('Distribution of Screen On Time (hours/day)')
plt.xlabel('Screen On Time (hours/day)')
plt.ylabel('Frequency')
```

```python
#scatter plot to find which age people use more screen time

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Screen On Time (hours/day)', data=df)
plt.title('Relationship between age and Screen On Time')
plt.xlabel('Age')
plt.ylabel('Screen On Time (hours/day)')
```

```python


# Line graph for age and screen time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='Screen On Time (hours/day)', data=df)
plt.title('Relationship between Age and Screen On Time')
plt.xlabel('Age')
plt.ylabel('Screen On Time (hours/day)')
plt.show()
```

```python
# displaying the correlation coefficients with a color scale to identify relationships between variables.
correlation_matrix = df.select_dtypes(include=np.number).corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#A histogram with a KDE (Kernel Density Estimate) overlay to visualize the distribution of  Screen On Time (hours/day).
sns.countplot(x='Age', data=df)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Categories by age to display a Screen time by Age
Age= df.groupby('Age')['Screen On Time (hours/day)'].mean()
Age.plot(kind='bar', figsize=(10, 6))
plt.title('Average Screen On Time By Age')
plt.xlabel('Age')
plt.ylabel('Average Screen On Time (Hours/hr)')
plt.show()
```

```python
#importing LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Train a simple linear regression model using Age as the predictor
X_age = df[["Age"]]  # Predictor variable (Age)
y_screen_time = df["Screen On Time (hours/day)"]  # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_age, y_screen_time, test_size=0.2, random_state=42)
```

```python
# Train the model
age_model = LinearRegression()
age_model.fit(X_train, y_train)
```

```python
# Predict screen time for twice the existing age
df["Predicted Screen Time (2x Age)"] = age_model.predict(df[["Age"]] * 2)
```
